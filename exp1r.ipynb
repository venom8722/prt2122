```r
library(dplyr)
library(ggplot2)
library(tidyr)
library(readr)

# Load dataset
df <- iris

# Basic stats
summary(df)

# Correlation only on numeric columns
cor(df[, 1:4])

# Tidy example
df_long <- pivot_longer(df, cols = 1:4, names_to = "Measurement", values_to = "Value")

# Visualization
ggplot(df_long, aes(x = Measurement, y = Value, fill = Species)) +
  geom_boxplot()
```
